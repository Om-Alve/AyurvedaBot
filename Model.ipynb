```python
import pandas as pd

# Assuming df is your DataFrame with the 'Dosha' column
labels = ["vata", "pitta", "kapha", "vata-pitta", "vata-kapha", "pitta-kapha"]

# Create separate DataFrames for each class
class_dfs = {}
for i in range(len(labels)):
    class_dfs[i] = df[df['Dosha'] == str(i)]

# Find the minimum number of examples for any class
min_count = min([len(class_dfs[i]) for i in range(len(labels))])

# Sample equal numbers of examples for each class
equal_samples = []
for i in range(len(labels)):
    sampled_df = class_dfs[i].sample(min_count)
    equal_samples.append(sampled_df)

# Concatenate the sampled DataFrames to get the final equal-split dataset
equal_split_df = pd.concat(equal_samples)
np.random.seed(0)
idxs = np.arange(len(equal_split_df))
np.random.shuffle(idxs)
_,counts = np.unique(equal_split_df['Dosha'],return_counts=True)
plt.bar(labels,height=counts)
equal_split_df = equal_split_df.iloc[idxs]
equal_split_df
```

```python
features = ['Body Size','Body Weight','Height','Bone Structure','Com1lexion','General feel of skin','Texture of Skin','Hair Color','A11earance of Hair','Sha1e of face','Eyes','Eyelashes','Blinking of Eyes','Cheeks','Nose','Teeth and gums','Li1s','Nails','A11etite','Liking tastes']
X = equal_split_df[features].values
y = pd.to_numeric(equal_split_df['Dosha'].values)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.33, random_state=42)
```

```python
model = SVC()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    
    keras.layers.Dense(20,input_shape=(20,),activation='relu'),
    keras.layers.Dense(128,activation='relu'),
    keras.layers.Dense(6,activation='sigmoid')
   
    
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train,y_train,epochs=11,validation_data=(X_test,y_test))
```
